```python
terror_d['nkill']=terror_d['nkill'].fillna(0)
terror_d['nwound']=terror_d['nwound'].fillna(0)
terror_d['Casualty']=terror_d['nkill']+terror_d['nwound']
terror_d.head()
```

```python
terror_d.describe()
```

```python
terror_d.head()
```

VISUALISING DATA

1. YEAR-WISE ATTACKS

```python
attacks=terror_d['iyear'].value_counts()
attacks
```

```python
attacks.plot(kind='bar',color='orange',figsize=(9,5),fontsize=14,width=0.1)
plt.xlabel('YEAR OF ATTACK',fontsize=14)
plt.ylabel('NUMBER OF ATTACKS',fontsize=14)
plt.title('YEAR-WISE ATTACKS',fontsize=14)
plt.show()
```

2. YEAR-WISE CASUALITIES

```python
sns.barplot(data=terror_d,x='iyear',y='Casualty',width=0.5)
```

3. DISPLAY OF ATTACKS ON THE BASIS OF REGION

```python
region=pd.crosstab(terror_d.iyear,terror_d.region_txt)
region.head()
```

```python
region.plot(kind='area',stacked='false',figsize=(10,5))
plt.title('REGION-WISE ATTACKS',fontsize=10)
plt.ylabel('NUMBER OF ATTACKS',fontsize=8)
plt.xlabel('YEAR OF ATTACK',fontsize=8)
plt.show()
```

4. COUNTRY-WISE ATTACKS

```python
cnt=terror_d['country_txt'].value_counts().head(15)
cnt
```

```python
cnt.plot(kind="bar",color='orange',figsize=(10,5),width=0.5)
plt.title('COUNTRY',fontsize=10)
plt.ylabel('NUMBER OF ATTACKS',fontsize=10)
plt.xlabel('COUNTRY-WISE ATTACK',fontsize=10)
plt.show()
```

5. GROUP-WISE ATTACKS

```python
grp_attack=terror_d['gname'].value_counts().head(20)
grp_attack
```

```python
grp_attack.plot(kind="bar",color='orange',figsize=(10,5),width=0.5)
plt.title('GROUP-WISE ATTACKS',fontsize=10)
plt.ylabel('NUMBER OF ATTACKS',fontsize=10)
plt.xlabel('GROUP',fontsize=10)
plt.show()
```

6. TARGET TYPE WISE ATTACKS

```python
target=terror_d['targtype1_txt'].value_counts().head(15)
target
```

```python
target.plot(kind="bar",color='orange',figsize=(10,5),width=0.5)
plt.title('TARGET TYPE-WISE ATTACKS',fontsize=10)
plt.ylabel('NUMBER OF ATTACKS',fontsize=10)
plt.xlabel('TARGET TYPE',fontsize=10)
plt.show()
```
