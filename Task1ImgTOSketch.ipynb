```python
import cv2
import matplotlib.pyplot as pl
pl.style.use('seaborn')
```

LOADING AN ORIGINAL IMAGE

```python
image=cv2.imread("cat.jpeg")
pl.figure(figsize=(8,8))
pl.imshow(image)
pl.axis("off")
pl.title("Oringinal")
pl.show()
```

CONVERTING IMAGE TO GRAYSCALE

```python
image_gray=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
pl.figure(figsize=(8,8))
pl.imshow(image_gray,cmap="gray")
pl.axis("off")
pl.title("Gray Scale Image")
pl.show()
```

INVERTING OF IMAGE

```python
image_invert=cv2.bitwise_not(image_gray)
pl.figure(figsize=(8,8))
pl.imshow(image_invert,cmap="gray")
pl.axis("off")
pl.title("Inverted Image")
pl.show()
```

SMOOTHING OF IMAGE

```python
img_smoothing=cv2.GaussianBlur(image_invert,(21,21),sigmaX=0,sigmaY=0)
pl.figure(figsize=(8,8))
pl.imshow(img_smoothing,cmap="gray")
pl.axis("off")
pl.title("Smoothen Image")
pl.show()
```

IMAGE TO PENCIL SKETCH

```python
fin=cv2.divide(image_gray,255-img_smoothing,scale=255)
pl.figure(figsize=(8,8))
pl.imshow(fin,cmap="gray")
pl.axis("off")
pl.title("Final Sketch Image")
pl.show()
```
